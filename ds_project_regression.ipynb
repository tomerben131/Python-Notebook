```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix , ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import Ridge
```

```python
df=pd.read_csv('Life Expectancy Data.csv')
df.head()
```

# **DATA CLEANING**

```python
df=df.drop(['Country','Year'],axis=1)
df.head()
#הורדתי את עמודות אלה כי הם לא רלוונטים לי לחישוב
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# prompt: can you fill all the missing slots using their median

# Fill missing values with the median for each column
for col in df.columns:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Verify that there are no more missing values
print(df.isnull().sum())
#כל המשתנים הם נומרים ולכן בחרתי להשתמש בכולם בחציון
```

```python
df.describe()
```

```python
df['Status']=df['Status'].map({'Developing':0,'Developed':1})
df.head()
```

```python
# prompt: put Life expectancy culomn as the last one

# Move the 'Life expectancy ' column to the last position
cols = list(df.columns)
cols.remove('Life expectancy ')
cols.append('Life expectancy ')
df = df[cols]
df.head()
```

```python
X=df.iloc[:,1:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,
                                                    y,
                                                    test_size=0.2,
                                                    random_state=0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# **linear trainig  and testing **






```python
# TRAIN MODEL using sklearn linearRegression
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)

# model evaluation for training set
y_test_predict = lin_model.predict(X_test)
Lin_r2=r2_score(Y_test, y_test_predict)
RMSE = root_mean_squared_error(Y_test, y_test_predict)
MAE=mean_absolute_error(Y_test, y_test_predict)
MAPE=mean_absolute_percentage_error(Y_test, y_test_predict)
print("The model performance for testing set")
print("--------------------------------------")
print('R squared:{}'.format(Lin_r2))
print('RMSE is {}'.format(RMSE))
print('MAE score is {}'.format(MAE))
print('MAPE score is {}'.format(MAPE))
print("\n")
```

```python
y_train_predict = lin_model.predict(X_train)
Lin_r2=r2_score(Y_train, y_train_predict)
RMSE = root_mean_squared_error(Y_train, y_train_predict)
MAE=mean_absolute_error(Y_train, y_train_predict)
MAPE=mean_absolute_percentage_error(Y_train, y_train_predict)
print("The model performance for training set")
print("--------------------------------------")
print('R squared:{}'.format(Lin_r2))
print('RMSE is {}'.format(RMSE))
print('MAE score is {}'.format(MAE))
print('MAPE score is {}'.format(MAPE))
print("\n")
```
